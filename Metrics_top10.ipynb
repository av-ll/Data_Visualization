```python
classification = data['Crime Index'].copy()
```

```python
class_row = []
for x in classification:
    if x <= 20:
        class_row.append('very low')
    elif 40 >= x > 20 :
        class_row.append('low')
    elif 60 >= x > 40 :
        class_row.append('moderate')
    elif 80 >= x > 60 :
        class_row.append('high')
    elif x > 80:
        class_row.append('very high')      
        
        
```

```python
dense = data[data['Density (per kmÂ²)']>600]
```

```python
dense[['Country','Crime Index','Rural % Population','Density (per kmÂ²)']].transpose()
```

```python
data['Crime_cat'] = class_row
```

```python
data['Area (kmÂ²)'].quantile(q=[0.2,0.4,0.6,0.8,0.90])
```

```python
denst = data['Area (kmÂ²)'].copy()
dens_row = []
for x in denst:
    if x <= 28266.0:
        dens_row.append('very small')
    elif 97337.2 >= x > 28266.0 :
        dens_row.append('small')
    elif 836874.4 >= x > 97337.2 :
        dens_row.append('medium')
    elif 1772192.8 >= x > 836874.4 :
        dens_row.append('big')
    elif x > 1772192.8:
        dens_row.append('very big')
    
```

```python
data['country_size_categorical'] = dens_row
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(11,6)})
freq_graph_colors = ['red','darkorange','gold','limegreen','forestgreen']
scatplot = sns.scatterplot(data =data,y='ppi',palette=freq_graph_colors,\
                           x='2019 HDI',hue='Crime_cat',size='country_size_categorical',size_order=\
                           ['very big','big','medium',\
                           'small','very small'])
scatplot.set_facecolor('xkcd:white')
scatplot.grid(color='lightgrey')
scatplot.set_xlim(xmin=0.4,xmax=1)
scatplot.legend(prop={'size': 9},ncol=2)
scatplot.set_title('Relation between PPI, HDI,Crime and Country Size',fontsize=20,fontweight='10')
plt.ylabel('PPI', size=10,rotation=0,weight='semibold')
plt.xlabel('2019 HDI', size=10,rotation=0,weight='semibold')

plt.savefig('4dim.png',dpi=200)
```
